```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def star_scatter(data, xname, yname, xindex, yindex, tittle, grid):
    plt.xlabel(xname, fontweight = "bold")
    plt.ylabel(yname, fontweight = "bold")
    plt.scatter(data[xindex], data[yindex], s = 2, c = "m")
    plt.title(tittle, fontweight = "bold")
    plt.grid(grid)
    plt.savefig("C:/Users/ASUS/GA_HW/HW04/" + tittle + ".png")
    plt.show()
    
def preprocess(file):
    with open(file, "r", encoding = "ASCII") as fh:
        list = fh.read()
        list = list.replace("  ", " ")
    with open(file, "w", encoding = "ASCII") as fh:
        fh.write(list)
    data = pd.read_csv(file, sep = " ", header = None)
    data.columns = range(1, len(data.columns) + 1)
    return data
```

```python
data = preprocess("summary.txt")
data
```

```python
star_scatter(data, "Age (years)", "Luminosity (log10(Lâ˜‰))", 2, 4, "(Fig.1)Luminosity to Age", True)
star_scatter(data, "Age (years)", "Radius (log10(Râ˜‰))", 2, 5, "(Fig.2)Radius to Age", True)
star_scatter(data, "Age (years)", "Surface temperature (log10(K))", 2, 6, "(Fig.3)Surface Temperature to Age", True)
star_scatter(data, "Age (years)", "Luminosity from PP chain (Lâ˜‰)", 2, 19, "(Fig.4)Luminosity from PP chain to Age", True)

total_luminosity = data[19] + data[20] + data[21] + data[22] + data[23]

plt.close()
plt.xlabel("Age (years)", fontweight = "bold")
plt.ylabel("Ratio of Luminosity from Specific Source(Lâ˜‰/Lâ˜‰total)", fontweight = "bold")
plt.grid(True)
plt.scatter(data[2], data[19] / total_luminosity, s = 0.5, c = "r")
plt.scatter(data[2], data[20] / total_luminosity, s = 0.5, c = "b")
plt.scatter(data[2], data[21] / total_luminosity, s = 0.5, c = "y")
plt.title("(Fig.5-1)Ratio of Luminosity from Specific Source to Age", fontweight = "bold")
plt.legend(["Lâ˜‰pp/Lâ˜‰total", "Lâ˜‰CNO/Lâ˜‰total", "Lâ˜‰triple/Lâ˜‰total"])
plt.savefig("C:/Users/ASUS/GA_HW/HW04/" + "(Fig.5-1)Ratio of Luminosity from Specific Source to Age" + ".png")
plt.show()

plt.close()
plt.xlabel("Age (years)", fontweight = "bold")
plt.ylabel("Ratio of Luminosity from Specific Source(Lâ˜‰/Lâ˜‰total)", fontweight = "bold")
plt.xlim(1.025 * 10 ** 10, 1.25 * 10 **10)
plt.grid(True)
plt.scatter(data[2], data[19] / total_luminosity, s = 0.5, c = "r")
plt.scatter(data[2], data[20] / total_luminosity, s = 0.5, c = "b")
plt.scatter(data[2], data[21] / total_luminosity, s = 0.5, c = "y")
plt.title("(Fig.5-2)Ratio of Luminosity from Specific Source to Age", fontweight = "bold")
plt.legend(["Lâ˜‰pp/Lâ˜‰total", "Lâ˜‰CNO/Lâ˜‰total", "Lâ˜‰triple/Lâ˜‰total"])

plt.savefig("C:/Users/ASUS/GA_HW/HW04/" + "(Fig.5-2)Ratio of Luminosity from Specific Source to Age" + ".png")
plt.show()
```
